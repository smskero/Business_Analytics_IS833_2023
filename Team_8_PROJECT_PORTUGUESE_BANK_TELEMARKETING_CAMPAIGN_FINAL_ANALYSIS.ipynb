**1.3 Dataset**

This data is sourced from "A Data-Driven Approach to Predict the Success of Bank Telemarketing" by Moro et. al., and is housed on the UCI Machine Learning Repository, an archive of datasets founded by UC Irvine students.

Overview of the attributes:

* ***Age:*** Age of customer
* ***Job:*** Job of customer
* ***Marital:*** Marital status of customer
* ***Education:*** Customer education level
* ***Default:*** Has credit in default?
* ***Housing:*** If costumer has housing loan
* ***Loan:*** Has Personal Loan
* ***Balance:*** Customer's individual balance
* ***Contact:*** Communication type
* ***Month:*** Last contact month of year
* ***Day:*** Last contact day of the week
* ***Duration:*** Last contact duration, in seconds
* ***Campaign:*** Number of contacts performed during this campaign and for this client
* ***PDays:*** Number of days that passed by after the client was last contacted from a previous campaign
* ***Previous:*** Number of contacts performed before this campaign and for this client
* ***POutcome:*** of the previous marketing campaign
* ***'Y':*** Has the client subscribed a term deposit

```python
import pandas as pd
from sklearn.preprocessing import OneHotEncoder, LabelEncoder

onehot_columns = ['education', 'month', 'contact','job','marital','poutcome']
onehot_encoder = OneHotEncoder()
onehot_encoded = onehot_encoder.fit_transform(df[onehot_columns]).toarray()

onehot_df = pd.DataFrame(onehot_encoded, columns=onehot_encoder.get_feature_names_out(onehot_columns))

label_columns = ['default', 'housing', 'loan', 'y']
label_encoder = LabelEncoder()
label_df = df[label_columns].apply(label_encoder.fit_transform)

df_encoded = pd.concat([df.drop(onehot_columns + label_columns, axis=1), onehot_df, label_df], axis=1)
```

Split the preprocessed data into training and testing sets.

But first split the data into two variables, where X represents the features (all columns except the target variable 'y') and y represents the target variable.

```python
from sklearn.model_selection import train_test_split

X = df_encoded.drop('y', axis=1)
y = df_encoded['y']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=9)
```

```python
##import pandas as pd
##from sklearn.preprocessing import StandardScaler

# Drop rows with string values from X_train
#X_train = X_train[X_train.applymap(lambda x: isinstance(x, (int, float)))]

# Create and apply StandardScaler object
#scaler = StandardScaler()
#X_train = scaler.fit_transform(X_train)
```

Then, standardize the feature data in the training and testing sets. This can ensure that all features are on the same scale and have a similar influence on the model's predictions, improving the model's performance and stability.

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

Class imbalance is a common problem in machine learning where one class (in this case, the minority class) is underrepresented compared to the other classes. This can lead to a biased model that performs poorly on the minority class. SMOTE (Synthetic Minority Over-sampling Technique) is a popular oversampling technique that generates synthetic examples of the minority class to balance the class distribution.

```python
from imblearn.over_sampling import SMOTE

smote = SMOTE(random_state=9)
X_train_resampled, y_train_resampled = smote.fit_resample(X_train, y_train)
```

```python

print("Class distribution before SMOTE:")
print(y_train.value_counts())
print("\nClass distribution after SMOTE:")
print(y_train_resampled.value_counts())
```

4.3. Selection of the model

a. Logistic Regression

```python
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import cross_val_score, KFold
from sklearn.metrics import roc_auc_score, precision_score, recall_score, f1_score

logreg = LogisticRegression(random_state=9, solver='lbfgs', max_iter=10000)
cv = KFold(n_splits=5, random_state=7, shuffle=True)
logreg.fit(X_train_resampled, y_train_resampled)

cv_scores = cross_val_score(logreg, X_train_resampled, y_train_resampled, cv=cv)
print(f'Logistic Regression k-Fold validation mean accuracy: {cv_scores.mean():.4f}')

y_pred = logreg.predict(X_test)

roc_auc = roc_auc_score(y_test, y_pred)
precision = precision_score(y_test, y_pred)
recall = recall_score(y_test, y_pred)
f1 = f1_score(y_test, y_pred)

print(f'Logistic Regression ROC AUC score: {roc_auc:.4f}')
print(f'Logistic Regression Precision: {precision:.4f}')
print(f'Logistic Regression Recall: {recall:.4f}')
print(f'Logistic Regression F1-score: {f1:.4f}')
print('-' * 50)
```
